```python
import pandas as pd
import numpy as np
```

```python
# city reductions
pop_arcgis = pd.read_csv('../../data/cities/city_pop_gdp_red/population_age_sex_arcgis.csv')
pop_arcgis.loc[pop_arcgis['ISO_CC'].isna(), 'ISO_CC'] = 'NA' # gets read as nan
pop_guppd = pd.read_csv('../../data/cities/city_pop_gdp_red/population_age_sex_guppd.csv')
gdp_arcgis = pd.read_csv('../../data/cities/city_pop_gdp_red/gdp_arcgis.csv')
gdp_arcgis.loc[gdp_arcgis['ISO_CC'].isna(), 'ISO_CC'] = 'NA' # gets read as nan
gdp_guppd = pd.read_csv('../../data/cities/city_pop_gdp_red/gdp_guppd.csv')
```

```python
# to match city with country names
regions = pd.read_csv('../../data/cities/city_pop_gdp_red/UNSD — Methodology.csv')
regions.loc[regions['ISO-alpha3 Code']=='NAM', 'ISO-alpha2 Code'] = 'NA' # gets read as nan
regions = regions[regions['ISO-alpha2 Code'].notnull()]
regions.shape
```

```python
# match cities with countries
pop_arcgis = pop_arcgis.merge(regions[['ISO-alpha2 Code', 'Country or Area']].rename(
    columns={'ISO-alpha2 Code': 'ISO_CC'}), how='left')
gdp_arcgis = gdp_arcgis.merge(regions[['ISO-alpha2 Code', 'Country or Area']].rename(
    columns={'ISO-alpha2 Code': 'ISO_CC'}), how='left')

pop_guppd = pop_guppd.merge(regions[['ISO-alpha3 Code', 'Country or Area']].rename(
    columns={'ISO-alpha3 Code': 'ISO'}), how='left')
gdp_guppd = gdp_guppd.merge(regions[['ISO-alpha3 Code', 'Country or Area']].rename(
    columns={'ISO-alpha3 Code': 'ISO'}), how='left')

# not in regions csv
pop_guppd.loc[pop_guppd['ISO']=='XKO', 'Country or Area'] = 'Kosovo' 
pop_guppd.loc[pop_guppd['ISO']=='TWN', 'Country or Area'] = 'Taiwan' 
gdp_guppd.loc[gdp_guppd['ISO']=='XKO', 'Country or Area'] = 'Kosovo' 
gdp_guppd.loc[gdp_guppd['ISO']=='TWN', 'Country or Area'] = 'Taiwan' 
```
